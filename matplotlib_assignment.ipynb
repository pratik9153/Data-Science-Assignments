```python
#2.

#A scatter plot is a diagram where each value in the data set is represented by a dot.

import matplotlib.pyplot as plt
import numpy as np

x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal (0,2,len(x))
plt.xlabel("this is x ")
plt.ylabel("this is y")
plt.title("normal distribution")
plt.scatter(x,y)
plt.show()
```

```python
#3.

#line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

#line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

#line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

#line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig,axs=plt.subplots(2,2)
axs[0,0].plot(x1,y1)
axs[0,0].set_title('Line 1')


axs[0,1].plot(x2,y2)
axs[0,1].set_title('Line 2')

axs[1,0].plot(x3,y3)
axs[1,0].set_title('Line 3')

axs[1,1].plot(x4,y4)
axs[1,1].set_title('Line 4')

plt.tight_layout()

plt.show()
```

```python
A bar plot in Matplotlib is a type of graph that represents categorical data with rectangular bars,
where the length of each bar corresponds to the value of the category it represents
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
plt.title("Growth")
plt.show()
```

```python
plt.barh(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
plt.title("Growth")
plt.show()
```
